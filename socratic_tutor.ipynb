```python
from openai import OpenAI

# Initialize client (replace with your API key)
client = OpenAI(api_key="your_api")

def socratic_tutor(user_input):
    prompt = f"""
You are a Socratic tutor. Do NOT give direct answers.

Your role is to guide the student by asking a sequence of thoughtful, step-by-step questions that lead them to discover the answer on their own.

Rules:
- Ask 5 to 8 questions
- Questions should be clear and progressively deeper
- Do NOT explain or answer
- Encourage reasoning

Concept/Question:
{user_input}

Now generate only the questions.
"""

    response = client.responses.create(
        model="gpt-5-mini",
        input=prompt
    )

    return response.output_text

# Test it
question = "What is photosynthesis?"
output = socratic_tutor(question)

print(output)
```

```python
while True:
    user_input = input("\nEnter a concept/question (or 'exit'): ")

    if user_input.lower() == "exit":
        break

    print("\nSocratic Tutor:\n")
    print(socratic_tutor(user_input))
```

With UI integration


```python
# Install required libraries
!pip install openai gradio pymupdf

import fitz  # PyMuPDF
import gradio as gr
from openai import OpenAI

# Add your OpenAI API Key
client = OpenAI(api_key="YOUR_API_KEY")

# Extract text from PDF
def extract_pdf_text(pdf_file):
    text = ""

    pdf_document = fitz.open(pdf_file.name)

    for page in pdf_document:
        text += page.get_text()

    return text

# Generate Socratic Questions
def socratic_tutor(pdf_file, topic):

    pdf_text = ""

    # If PDF uploaded
    if pdf_file is not None:
        pdf_text = extract_pdf_text(pdf_file)

    prompt = f"""
You are a Socratic tutor.

Your job is to guide learning using ONLY questions.

Do NOT give answers.
Do NOT explain concepts directly.

Ask 5 to 8 thoughtful questions that help the learner discover the answer on their own.

Topic:
{topic}

Reference Material:
{pdf_text[:4000]}

Generate only the questions.
"""

    response = client.responses.create(
        model="gpt-5-mini",
        input=prompt
    )

    return response.output_text

# Gradio Interface
with gr.Blocks(theme=gr.themes.Soft()) as app:

    gr.Markdown("""
    # 🧠 Socratic Tutor AI

    Upload a PDF or enter a topic.

    The AI will teach by asking guiding questions instead of giving direct answers.
    """)

    with gr.Row():

        pdf_input = gr.File(
            label="Upload PDF (Optional)",
            file_types=[".pdf"]
        )

        topic_input = gr.Textbox(
            label="Enter Topic or Question",
            placeholder="Example: What is photosynthesis?"
        )

    generate_btn = gr.Button("Generate Socratic Questions")

    output_box = gr.Textbox(
        label="AI Tutor Output",
        lines=15
    )

    generate_btn.click(
        fn=socratic_tutor,
        inputs=[pdf_input, topic_input],
        outputs=output_box
    )

# Launch app
app.launch()
```
